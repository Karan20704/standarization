```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('K:\Machine learning campus x\\Standarization\\Social_Network_Ads.csv' , usecols=['Age' , 'EstimatedSalary' , 'Purchased'])
df.head()
```

```python
df.describe()
```

```python
import pandas as pd

df1 = pd.DataFrame({
    "Age": [1, 2, 1],
    "EstimatedSalary": [300000, 200000, 3000],
    "Purchased": [0, 1, 0]
})
df1.head()
```

```python
df = pd.concat([df, df1], ignore_index=True,axis=0)
df.head()
```

```python
df.tail()
```

```python
sns.scatterplot(data=df, x='Age', y='EstimatedSalary', hue = 'Purchased')
plt.title('Data With Outliers')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x = df.iloc[:,:-1]
y = df.iloc[:,-1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)   
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
sns.scatterplot(x = x_train['Age'], y = x_train['EstimatedSalary'], hue = y_train)
plt.title('Train Data Before Scaling')
plt.show()
```

```python
sns.scatterplot(x = x_train_scaled[:,0], y = x_train_scaled[:,1], hue = y_train)
plt.title('Train Data After Scaling')
plt.show()  
```

```python
# using standard scaler outlier effect is visible
```

## robust scaling

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
sns.scatterplot(x = x_train_scaled[:,0], y = x_train_scaled[:,1], hue = y_train)
```

```python
#  The main cluster of data gets scaled tightly
#  Extreme outliers dont strongly influence the scale
#  But outliers still remain  just mapped far from main cluster
```

```python
# in short help to avoid an outlier effect
```
